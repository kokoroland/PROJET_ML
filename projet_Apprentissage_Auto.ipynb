# What Hosts And Areas are telling us?

Data posses the information about 11452 different hosts who rented their property in New York City.

They are located at different locations in the city itslef.

There are 47906 unique properties listed by airbnb which belongs to 11452 different hosts. Top 5 hosts having largest number of properties include -Michael - 417 David - 403 Sonder (NYC) - 327 John - 294 and Alex - 279

All properties are located within five neighbourhood groups in New York City. i.e 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island', 'Bronx'.

There are three types of options available for guests to stay-Entire home or apartment, Private room,Shared room. Availability of each segment is as follows.

Entire home/apt - 25409
Private room -22326
Shared room - 1160

Most Famous Property in terms of total number of reviews is 'Room near JFK Queen Bed' owned by Dona, situated at Queens Neighborhood group.Runner ups are- Great Bedroom in Manhatten and Beautiful Bedroom in Manhattan.

Maximum number of properties located in ings are in Williamsburg neighbourhood in Brooklyn neighbourhood group.

Minimum number of properties located in Willowbrook-1,New Dorp-1,Richmondtown-1,Rossville-1.(One in each neighbourhood)

```python
sns.countplot(x = df['neighbourhood_group'], palette="Paired")
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Neighbourhood Group')
```

Nous pouvons voir que le nom le plus eleve d'enregistrement de Airbnb est <strong>Manhattan</strong>
c'est est peut la raison pour laquelle est <strong>Manhattan</strong> est celebre en terme d'attraction des tourists et ces centres d'affaires et d'evenements culturels


```python
sns.barplot(x ='neighbourhood_group',y = 'price' ,hue = 'room_type' ,data = df, palette ="magma")
plt.figure(figsize=(10, 10))
```

Nous voyons que le prix est eleve a  <strong>Manhattan</strong> pour les trois type de chambre
et la chambre <strong>" shared room"</strong> est plus accessible a <strong>Brooklyn</strong>



```python
top10_host = df.host_id.value_counts().head(10).index
top10_host_vis=df.loc[df['host_id'].isin(top10_host)]
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data=top10_host_vis,x='neighbourhood',order = top10_host_vis['neighbourhood'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Top 10 hosts neighbourhood",weight='bold')
plt.ylabel('Number of Airbnb')
```

 We're seeing more airbnb homes in the financial district and midtown.
 We can assume that this is directly proportional to the fact that people who come for business or touristic trips prefer these regions more.

```python
avaibility = df[df.price < 500].plot(kind='scatter', x='longitude', y='latitude', label='availability_365', c='price',
                  cmap=plt.get_cmap('jet'), colorbar=True, alpha=0.4, figsize=(10,8))
avaibility.legend()
```

 We see that house prices are higher in Manhattan and some inland areas and along the ocean.
We know that prices are high, as Manhattan is famous for hosting many tourist attractions, business centers, and cultural events.

The high prices in the interior may be associated with the central location of touristic places and financial institutions 
 and the better transportation network.

 It can be deduced that the landscape and nature factors are also effective in the price increase on the ocean sides.

 Regardless of all these conditions, it is important to remember that whether the house is bigger or new can also determine the price.
